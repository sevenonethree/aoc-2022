```python
import pandas as pd
```

```python
with open('./input.txt','r') as f:
    lines = f.readlines()
```

## Input information
* A/X = Rock
* B/Y = Paper
* C/Z = Scissors

|Opponent| Me |Choice Score | Total Score (+6 for win, +3 for tie)|
|---|---|---|---|
|A | Y | 2 | 8 |
|B | X | 1 | 1 |
|C | Z | 3 | 6 |

* In the first round, your opponent will choose Rock (A), and you should choose Paper (Y). This ends in a win for you with a score of 8 (2 because you chose Paper + 6 because you won).
* In the second round, your opponent will choose Paper (B), and you should choose Rock (X). This ends in a loss for you with a score of 1 (1 + 0).
* The third round is a draw with both players choosing Scissors, giving you a score of 3 + 3 = 6.


```python
scoreDict = {
      'AY': 8,
      'BZ': 9,
      'CX': 7,
      'AX': 4,
      'BY': 5,
      'CZ': 6,
      'BX': 1,
      'CY': 2,
      'AZ': 3,
}
```

```python
chal1guide = []
for line in lines:
    cleaned = line.replace('\n', '').replace(' ', '')
    chal1guide.append([cleaned, scoreDict[cleaned]])
#     
#     guide.append(cleaned)
chal1guide
```

```python
df = pd.DataFrame(chal1guide, columns=['prediction', 'score'])
df
```
