```python
with open('./input.txt','r') as f:
    lines = f.readlines()

lines = [line.replace('\n', '') for line in lines]
```

## Challenge 1

```python
unwieldyDict = {}
# chr(97) lower a
# insert these first, because it makes more sense in my mind since they have higher priority
for priority in range(26):
    unwieldyDict[chr(97 + priority)] = priority + 1
# chr(65) upper A
for priority in range(26):
    unwieldyDict[chr(65 + priority)] = priority + 27
```

```python
for pack in lines:
    packSize = len(pack)
    compartment1 = pack[:packSize//2]
    compartment2 = pack[packSize//2:]
    print(pack)
    print(compartment1)
    print(compartment2)
    print(sorted(compartment1))
    print(sorted(compartment2))
    print('\n')

# packSize = len(lines[0].replace('\n',''))

# compartment1 = lines[0][:packSize//2]
# compartment2 = lines[0][packSize//2:]

```

```python
def get_mispacks():
    mispacks = []
    for pack in lines:
        packSize = len(pack)
        compartment1 = sorted(pack[:packSize//2])
        compartment2 = sorted(pack[packSize//2:])
        common = [item for item in compartment1 if item in compartment2]
        mispacks.append(common[0])
    return mispacks

mispacks = get_mispacks()
mispackValues = []

for screwup in mispacks:
    mispackValues.append(unwieldyDict[screwup])

sum(mispackValues)
```

## Challenge 2

```python
def get_common_items(group_members):
    pack1 = sorted(group_members[0])
    pack2 = sorted(group_members[1])
    pack3 = sorted(group_members[2])
    common = list(set([item for item in pack1 if item in pack2 and item in pack3]))
#     print(common)
    return common

def generate_groups():
    groups = {}
    currentGroup = []

    for idx, x in enumerate(lines):
        if (idx-3)%3 == 0:
            # start new group
            currentGroup =(lines[idx:idx+3])
        else:
            if len(currentGroup) > 0:
                groups[f'group{(idx//3)+1}'] = currentGroup
            currentGroup = []    
    return groups

elf_groups = generate_groups()

group_names = []

for g in elf_groups:
#     print(elf_groups[g],'\n\n')
    [group_names.append(item) for item in get_common_items(elf_groups[g])]

group_values = [unwieldyDict[value] for value in group_names]
sum(group_values)
```
