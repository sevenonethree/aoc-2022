```python
import pandas as pd
```

```python
with open('./input.txt','r') as f:
    lines = f.readlines()
```

```python
elfNumber = 1
elfItems = 0
elfSum = 0
elves = []
# elf = [id, items, totalCalories]
for line in lines:
    if line == '\n':
        # print(f'end of elf {elfNumber}')
        elves.append([elfNumber, elfItems, elfSum])
        elfNumber += 1
        elfItems = 0
        elfSum = 0
    # print(str.replace(line,'\n',''))
    else:
        elfItems += 1
        elfSum += int(str.replace(line,'\n', ''))

elves
```

```python
df = pd.DataFrame(elves, columns=['elfId', 'itemCount', 'totalCalories'])
```
