```python
def generate_run(lines):
    history = []
    update_on = 0
    cycle = 1
    next_cycle = cycle
    reg = 1
    for line in lines:
        instr = line.split()
        next_cycle = len(instr)

        if len(instr) > 1: 
            reg += int(instr[1])

        # print(f'current cycle: {cycle}\ninstruction: {line}\nupdate register to: {reg}\nupdate on: {cycle + next_cycle}\n')
        history.append({
            'cycle': cycle,
            'reg': reg,
            'update': cycle + next_cycle,
            'instruction': line
        })
        cycle += next_cycle
    return history
```
